```python
def find_equal_sum_columns(df, row_num):
    """table의 크기를 출력하는 함수.

    :param df: 데이터프레임
    :param row_num: 특정 행 번호
    :return: table의 크기
    """
    prev_sum = 0    #이전까지의 열방향의 합을 저장할 변수
    count = 0   # 더한 열의 개수를 저장할 변수
    
    # row_num번째 행의 모든 열을 순회하며 비교
    for col_name in range(2,len(df.columns)):
        curr_val = pd.to_numeric(df.iloc[row_num, col_name])

        #셀이 나머지 칸의 합일 경우 count를 return
        if prev_sum == curr_val and curr_val != 0:
            return count
        
        # 이전까지의 열방향의 합과 현재 열의 값이 다른 경우
        else:
            prev_sum += curr_val
            count += 1
    
    # 모든 열을 순회하면서 조건에 맞는 값을 찾지 못한 경우
    return count
```

```python
import pandas as pd
import os 
import time
from google.colab import files
import io
import openpyxl

#현재 시간 저장
now = time.strftime('%Y%m%d%H%M%S')

#directory 새로 만들고 바꾸기
os.makedirs('/content/{}'.format(now))
os.chdir('/content/{}'.format(now))

#전처리할 파일을 업로드 하고 sheet number을 입력해주세요.
uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_excel(filename, sheet_name=int(input("choose sheetnumber to refine >>>")), header= 5)
df.iloc[:,0] = df.iloc[:,0].fillna(0).astype(int)
```

```python
# 전체 csv파일의 크기를 n_row와 n_col에 저장
n_row = df.shape[0]
n_col = df.shape[1]

size_x = find_equal_sum_columns(df, 0)      # matrix X의 사이즈

print(f'size of matrix is {size_x}')
new_df = df.iloc[0:size_x, 0:size_x]
new_df.to_csv(f'matrixX_{size_x}_by_{size_x}.csv', index=False)

# 나머지 행과 열들을 각각의 csv파일로 저장
for i in range(size_x, n_row):
    label = str(df.iloc[i, 1]).replace("\n", "")
    df.iloc[i, 2:n_col].to_csv(f'j{i+1-size_x}_{label}.csv', index=False, header=[label])

for i in range(size_x, n_col):
    label = str(df.columns[i]).replace("\n", "")
    df.iloc[0:n_row, i].to_csv(f'k{i+1-size_x}_{label}.csv', index=False, header=[label])
```
